```python
import pandas as pd
```

```python
FILEPATH = "../countries-of-the-world.csv"
```

```python
df = pd.read_csv(FILEPATH)
```

```python
df.head(2)
```

```python
for col in df.columns:
    print(col)
```

```python
# print all cols

new_cols = {}
for old_col in list(df.columns):
    # print(col)

    col = old_col
    col = col.lower()

    col = col.replace(".", "")
    col = col.replace(" ", "_")
    col = col.replace("(", "")
    col = col.replace(")", "")
    col = col.replace("%", "")
    col = col.replace("/", "")
    col = col.replace("$", "")
    
    # if " " in col or "(" in col:
    # print(col)

    new_cols[old_col] = col
```

```python
new_cols
```

```python
df.rename(
    columns = new_cols, 
    inplace = True
)
```

```python
df.head(2)
```

```python
df.to_csv('../countries-of-the-world-cleaned.csv')
```
