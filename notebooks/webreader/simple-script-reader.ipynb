```python
import requests
from bs4 import BeautifulSoup
```

```python
# Collect and parse first page
page = requests.get('https://www.springfieldspringfield.co.uk/view_episode_scripts.php?tv-show=two-and-a-half-men&episode=s01e01')
soup = BeautifulSoup(page.text, 'html.parser')  
```

```python
content = soup.select('div.scrolling-script-container')[0]

contents = content.children  
```

```python
lines = []
dcounter = 0
```

```python
for c in contents:

    dcounter += 1    

    #print(dcounter, c)
    if(c is None):
        continue

    #print(type(c))
    try:
        c = c.replace('<br/>', '')
        c = c.strip()
        lines.append(c)            
    except TypeError as e:
        #print('s error :')
        pass 
```

```python
counter = 0
for line in lines:
    print(counter, line)
    counter += 1
```
