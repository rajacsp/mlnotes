```python
from io import BytesIO
import requests
import pandas as pd
```

```python
filename = 'https://gitlab.com/rajacsp/datasets/raw/master/amazon_co_ecommerce_sample.csv'

r = requests.get(filename)
data = r.content
```

```python
df = pd.read_csv(BytesIO(data), index_col=0)
```
