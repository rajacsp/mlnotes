```python
import numpy as np
```

```python
a = np.array([1, 4, 5, 6])
```

```python
print(a)
```

```python
a[0] = 4
```

```python
print(a)
```

```python
a.flags.writeable = False
```

```python
print(a)
```

```python
a[2] = 3
```

##### Rule:

You can't change the value of a numpy array after it is marked with writeable False. 

```python
a.flags.writeable = True
```

```python
a[2] = 8
```

```python
print(a)
```

```python
# flags.writeable = True will enable them to be updated
```
