```python
from functools import reduce
```

```python
def do_magic(a, b): 
    
    if((a % 2 == 1) & (b % 2 == 1) ):
        return a * b
    
    return a + b
```

```python
do_magic(10, 15)
```

```python
do_magic(11, 15)
```

```python
reduce(do_magic, [10, 20, 30, 40])
```

```python
reduce(do_magic, [10, 20, 30, 40, 31, 41])
```
