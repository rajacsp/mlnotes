```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-5, 5, 1)
y = np.arange(-5, 5, 1)

x = np.array([1, 2, 3, 4])
y = np.array([9, 10, 30, 10])

xx, yy = np.meshgrid(x, y, sparse=True)

z = np.sin(xx**2 + yy**2) / (xx**2 + yy**2)

plt.contourf(x,y,z)
plt.show()
```
