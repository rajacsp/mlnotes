```python
import numpy as np
```

```python
mat_a = np.mat('1 2; 3 4')
mat_b = np.mat('5 6; 7 8')
```

```python
mat_a
```

```python
ar_a = np.array([
    [1, 2],
    [3, 4]
])

ar_b = np.array([
    [5, 6],
    [7, 8]
])
```

```python
ar_a
```

```python
# Matrix multiplication

mat_c = mat_a * mat_b
```

```python
mat_c
```

```python
# Numpy array multiplication (element wise)

ar_c = ar_a * ar_b
```

```python
ar_c
```

```python
# Numpy dot multiplication

ar_d = np.dot(ar_a, ar_b)
```

```python
# We can also use @
ar_d = ar_a @ ar_b
```

```python
ar_d
```
