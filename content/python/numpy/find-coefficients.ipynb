```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 2, 3, 4, 4])
y = np.array([2, 4, 6, 8, 9])
```

```python
print(x)
```

```python
print(y)
```

```python
size = np.size(x)
```

```python
x_mean = np.mean(x)
```

```python
print(x_mean)
```

```python
y_mean = np.mean(y)
```

```python
print(y_mean)
```

```python
SS_xy = np.sum(y * x - (size * y_mean * x_mean))
```

```python
print(SS_xy)
```

```python
SS_xx = np.sum(x * x - (size * x_mean * x_mean))
```

```python
print(SS_xx)
```

```python
# Calculate regression coefficients
a1 = SS_xy / SS_xx
b1 = y_mean - (a1 * x_mean)
```

```python
coeff = b1, a1
```

```python
print(coeff)
```
