```python
import numpy as np
```

```python
a = np.array([1, 2])
```

```python
b = np.array([
    [1, 2, 3],
    [4, 5, 6]
])
```

```python
a
```

```python
b
```

```python
(a[:, np.newaxis] * b)
```

```python
c = (a[:, np.newaxis] * b).sum(axis=1)
```

```python
c
```

```python
c_einsum = np.einsum('i, ij->i', a, b)
```

```python
c_einsum
```
